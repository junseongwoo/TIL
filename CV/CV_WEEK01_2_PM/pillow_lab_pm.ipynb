오후에 이미지 메타 정보를 다루는 것을 해보았다. 


JPG 파일에는 Exchangeable Image File Format (Exif)에 메타정보를 가지고 있다.
Exif : 카메라의 설정 (ISO, 초점, 밝기 등등), 위치 정보 

Tag Image File Format (Tiff) 메타정보를 가지고도 있다.

Exif Tag
GPS Exif
Tiff Tag

```python
from PIL import Image
from PIL.ExifTags import TAGS

def get_exif(image_file_path):
    exif_table = {}
    image = Image.open(image_file_path)
    info = image.getexif()
    for tag, value in info.items():
        decoded = TAGS.get(tag, tag)
        exif_table[decoded] = value
    return exif_table

if __name__ == "__main__":
    exif = get_exif("bridge.JPG")
    print(exif)
```

```python
from PIL import Image
from PIL.ExifTags import TAGS, GPSTAGS

def get_exif(image_file_path):
    exif_table = {}
    image = Image.open(image_file_path)
    info = image.getexif()
    for tag, value in info.items():
        decoded = TAGS.get(tag, tag)
        exif_table[decoded] = value
    return exif_table
    
    gps_info = {}
    for key in exif_table["GPSInfo"]:
        decode = GPSTAGS.get(key, key)
        gps_info[decode] = exif_table['GPSInfo'][key]

    if gps_info:
        return gps_info
    else:
        return "GPS 데이터가 존재하지 않습니다."

if __name__ == "__main__":
    exif = get_exif("jester.jpg")
    print(exif)
```

```python
from PIL import Image 
from PIL.TiffTags import TAGS

def get_metadata(image_file_path):
    image = Image.open(image_file_path)
    metadata = {}
    for tag in image.tag.items():
        metadata[TAGS.get(tag[0])] = tag[1]
    return metadata

if __name__ == "__main__":
    metadata = get_metadata("MARBIBM.TIF")
    print(metadata)
```
