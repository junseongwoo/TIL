### 데이터 리스트에 저장
- 최고 기온 데이터를 날짜 순으로 저장 
- result라는 빈 리스트를 만들어 저장

```python
f = open('../Data_Analysis_for_Everyone/data/seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

print(result)
```

```python
# split() 함수로 8월의 최고 기온 데이터를 추출
f = open('../Data_Analysis_for_Everyone/data/seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.show()
```

```python
# 4월 30일, 생일을 기준으로 온도 추출 
f = open('../Data_Analysis_for_Everyone/data/seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '04' and \
            row[0].split('-')[2] == '30':
            result.append(float(row[-1]))

plt.plot(result, 'blue')
plt.show()
```

```python
f = open('../Data_Analysis_for_Everyone/data/seoul.csv')
data = csv.reader(f)
next(data)

high = [] 
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-') 

        if 1993 <= int(date[0]):
            if date[1] == '04' and date[2] =='30':
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.title('내 생일 기온 변화 그래프')
plt.plot(high, 'red', label = 'high')
plt.plot(low, 'blue', label = 'low')
plt.legend() 
plt.show()
```
