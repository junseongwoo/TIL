### title() 함수로 그래프에 제목 추가

```python
plt.title('plotting')
plt.plot([10, 20, 30, 40])
plt.show()
```

### 그래프에 범례 추가
- 범례는 보통 두 개 이상의 데이터를 표시할 때 사용 
- plot() 함수에 label이라는 속성의 레이블 값으로 원하는 문자열을 넣어줌
- 그래프 그리기 전에 legend() 함수를 실행시키면 레이블 값이 범례로 나타남
- 범례 위치 legend(loc = 숫자)

|2|9|1|
|:---:|:---:|:---:|
|6|10|5,7|
|3|8|4|

```python
plt.title('legend')
plt.plot([10, 20, 30, 40], label = 'asc')
plt.plot([40, 30, 20, 10], label = 'desc')
plt.legend()
plt.show()
```

### 그래프 색상 바꾸기
- color 속성을 추가하여 색상 바꿈
- color 속성을 생략하면 자동으로 색상 설정

```python
plt.title('color')
plt.plot([10, 20, 30, 40], color = 'skyblue', label = 'skyblue')
plt.plot([40, 30, 20, 10], 'pink', label = 'pink')
plt.legend(loc = 9)
plt.show()
```

### 그래프 선 모양 바꾸기 
- plot() 함수의 linestyle 속성에 원하는 선 모양 지정 
- linestyle 대신 ls라고 작성 가능

```python
plt.title('line style')
plt.plot([10, 20, 30, 40], color = 'b', label = 'dashed', linestyle = '--')
plt.plot([40, 30, 20, 10], color = 'g', label = 'dotted', ls = ':')
plt.plot([25, 25, 25, 25], 'r--', label = 'color ls')
plt.legend(loc = 3)
plt.show()
```

### 마커 모양 바꾸기
- plot() 함수에 marker 속성을 설정하여 점 형태로 그래프를 그림 
- 색상과 marker 모양을 한번에 설정 가능 

```python
plt.title('marker')
plt.plot([10, 20, 30, 40], 'b.', label = 'circle')
plt.plot([40, 30, 20, 10], 'g^', label = 'triangle up')
plt.plot([25, 25, 25, 25], 'r.--', label = 'color marker ls')
plt.legend()
plt.show()
```
