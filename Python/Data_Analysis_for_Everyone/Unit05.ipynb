### 기온 데이터를 히스토그램으로 표현

```python
import csv
f = open('../Data_Analysis_for_Everyone/data/seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))

plt.hist(result, bins = 100, color = 'r')
plt.show()
```

```python
# 8월 데이터만 그래프 그리기
f = open('../Data_Analysis_for_Everyone/data/seoul.csv')
data = csv.reader(f)
next(data)
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))

plt.hist(aug, bins = 100, color = 'r')
plt.show()
```

```python
# 1월과 8월 데이터 히스토그램 시각화 
f = open('../Data_Analysis_for_Everyone/data/seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))

        if month == '01':
            jan.append(float(row[-1]))


plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```

### 기온 데이터를 상자 그림으로 표현 
- boxplot 형태로 시각화 
- boxplot은 가공하지 않은 자료를 그대로 이용하는 것이 아니라, 
- 자료에서 얻어낸 최댓값, 최솟값, 상위 1/4, 2/4(중앙), 3/4에 위치한 값을 보여주는 그래프 

```python
result = [] 
for i in range(13):
    result.append(random.randint(1, 1000))

plt.boxplot(result)
plt.show()
```
