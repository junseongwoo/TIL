### 인구 구조를 다양한 형태로 시각화

```python
import csv
import matplotlib.pyplot as plt

f = open('../Data_Analysis_for_Everyone/data/age.csv')
data = csv.reader(f)
result = []

name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:]: 
            result.append(int(i.replace(',','')))

plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name)
plt.plot(result)
plt.show()
```
