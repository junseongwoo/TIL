### 그래프를 그림 파일로 저장
- matplotlib 라이브러리로 생성한 그래프를 파일로 저장하는 방법
- savefig() 함수에 저장하려는 파일의 이름과 파일 경로 설정 
- transparent= True 옵션으로 그림 배경을 투명하게 지정
- 색상을 정하는 컬러맵(cmap)으로 'viridis' 옵션 사용

```python
import pandas as pd 
import matplotlib.pyplot as plt 

plt.style.use('default')

path = 'C:\\Users\\jswoo\\Desktop\\Repository\\Study_Data_Science\\Pandas_Data_Analytics\\Part04\\auto-mpg.csv'
df = pd.read_csv(path, header = None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
'acceleration', 'model year', 'origin', 'name']
```

```python
cylinders_size = df.cylinders/df.cylinders.max() * 300

df.plot(kind = 'scatter', x = 'weight', y = 'mpg', marker = '+', figsize = (10,5), 
cmap = 'viridis', c = cylinders_size, s = 50, alpha = 0.3)

plt.title('Scatter Plot : mpg-weight-cylinders')

plt.savefig("C:\\Users\\jswoo\\Desktop\\Repository\\Study_Data_Science\\Pandas_Data_Analytics\\Part04\\scatter.png")
plt.savefig("C:\\Users\\jswoo\\Desktop\\Repository\\Study_Data_Science\\Pandas_Data_Analytics\\Part04\\scatter_transparent.png", transparent = True)
plt.show()
```
