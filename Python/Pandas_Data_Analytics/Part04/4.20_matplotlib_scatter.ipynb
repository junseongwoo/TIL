### 산점도 

- 서로 다른 두 변수 사이의 관계를 나타냄 
- 각 변수는 연속되는 값을 가지며 일반적으로 int64또는 float64 값이다.
- plot() 메소드에 kind = 'scatter' 옵션으로 산점도를 그린다.

```python
import pandas as pd 
import matplotlib.pyplot as plt 

plt.style.use('default')

path = 'C:\\Users\\jswoo\\Desktop\\Repository\\Study_Data_Science\\Pandas_Data_Analytics\\Part04\\auto-mpg.csv'
df = pd.read_csv(path, header = None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
'acceleration', 'model year', 'origin', 'name']
```

```python
df.plot(kind = 'scatter', x = 'weight', y = 'mpg', c = 'coral', 
s = 10, figsize = (10, 5))
plt.title('Scatter Plot - mpg vs weight')
plt.show()
```
