### 히스토그램

- 변수가 하나인 단변수 데이터의 빈도수를 그래프로 표현
- x축을 같은 크그의 여러 구간으로 나누고 각 구간에 속하는 데이터 값의 개수를 y축에 표시
- plot() 메소드에 kind = 'hist' 옵션을 설정
- bins = 10을 지정하여 10개 구간으로 나눔 

```python
import pandas as pd 
import matplotlib.pyplot as plt 

plt.style.use('classic')

path = 'C:\\Users\\jswoo\\Desktop\\Repository\\Study_Data_Science\\Pandas_Data_Analytics\\Part04\\auto-mpg.csv'
df = pd.read_csv(path, header = None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
'acceleration', 'model year', 'origin', 'name']

df['mpg'].plot(kind='hist', bins = 10, color = 'coral', figsize = (10,5))

plt.title('HIstogram')
plt.xlabel('mpg')
plt.show()
```
