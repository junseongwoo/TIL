### 불린 인덱싱 
- series 객체에 어떤 조건식을 적용하면 각 원소에 대해 불린 값으로 구성된 시리즈 반환 
- 참에 해당하는 데이터 값을 따로 선택 가능 -> 어떤 조건을 만족하는 데이터만을 추출
- df의 각 열은 series 객체이므로, 조건식을 적용하면 불린 시리즈를 만들 수 있음
- df의 불린 인덱싱 : DataFrame 객체[불린 시리즈]

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')

mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic.loc[mask1, :]
df_teenage.head()
```

- 서로 다른 두 열에 조건식 적용

```python
mask2 = (titanic.age < 10) & (titanic.sex == 'female')
df_female_under10 = titanic.loc[mask2, :]
df_female_under10
```

- 두 조건식을 OR(|) 연산자로 결합하여 두 조건 중에서 하나라도 참인 값을 추출 


```python
mask3 = (titanic.age < 10) | (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3, ['age', 'sex', 'alone']]
df_under10_morethan60.head()
```
