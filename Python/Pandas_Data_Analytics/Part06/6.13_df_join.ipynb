### 데이터프레임 결합
- pandas join() 메소드는 두 df의 행 인덱스를 기준으로 결합
- on = keys 옵션을 설정하면 행 인덱스 대신 다른 열을 기준으로 결합 가능 
- 행 인덱스 기준으로 결합 : DataFrmae.join(DataFrame2, how = 'left')

```python
import pandas as pd 

# 디스플레이 설정 변경 
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_colwidth', 20)
pd.set_option('display.unicode.east_asian_width', True)

path1 = '../data/stock price.xlsx'
path2 = '../data/stock valuation.xlsx'

df1 = pd.read_excel(path1, engine='openpyxl', index_col = 'id')
df2 = pd.read_excel(path2, engine='openpyxl', index_col = 'id')

df3 = df1.join(df2)
df3
```

- how = 'inner' 옵션을 적용하여 두 df에 공통으로 존재하는 행 인덱스를 기준으로 추출 
- 'id' 열을 기준으로 df1과 df2에 공통으로 속하는 종목 찾아냄

```python
df4 = df1.join(df2, how = 'inner')
df4
```
