### 날짜 데이터 분리
- 연-월-일 날짜 데이터에서 일부를 분리하여 추출
- 날짜 정보가 있는 열을 to_datetime() 함수에 전달하여 Timestamp로 변환
- 변환된 결과를 새로운 열에 담에 데이터프레임에 추가
- dt 속성을 이용하여 연, 월, 일을 개별적으로 추출 
- 추출한 데이터를 새로운 열에 저장하고 데이터프레임에 추가

```python
import pandas as pd 

path = 'C:\\Users\\jswoo\\Desktop\\Repository\\Study_Data_Science\\Pandas_Data_Analytics\\stock-data.csv'
df = pd.read_csv(path)
```

```python
df['new_date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df['Year'] = df['new_date'].dt.year
df['Month'] = df['new_date'].dt.month
df['Day'] = df['new_date'].dt.day

df.head()
```

- to_period() 메소드를 적용하여 연-월 또는 연도를 추출 

```python
df['Date_yr'] = df['new_date'].dt.to_period(freq='A')
df['Date_m'] = df['new_date'].dt.to_period(freq='M')

df.head()
```

- 추출한 날짜 정보를 데이터프레임의 행 인덱스로 지정

```python
df.set_index('Date_m', inplace = True)
df.head()
```
